```python
import pandas as pd 
import numpy as np 
import plotly.express as px
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# I download the top canadian companies ticker symbols info using the download method/function
multi_tickers=yf.download("RY,CP,TD",start="2018-01-01",end="2023-04-27",group_by="tickers")
```

```python
# we can view the information in history like format below
multi_tickers
```

```python
#I use the max function to findout which company has the highest prices 
multi_tickers.max()
```

```python
# we use the min function to determie the lowest Prices of each of these top company Giants 
multi_tickers.min()
```

```python
# we find the average prices for each Company 
multi_tickers.mean()
```

```python
# use the mode to analyse the companies information 

multi_tickers.mode
```

```python
#I use std to find the standard deviation of each company
multi_tickers.std()
```

```python
multi_tickers.std()
```

```python
fig,axes=plt.subplots(2,2,figsize=(25,25))
multi_tickers.plot(ax=axes[0,0],title='A Dashboard Showing All Canadian Companies Maximum Price Information')
multi_tickers.max().plot(ax=axes[0,1],kind="bar")
multi_tickers.max().plot(ax=axes[1,0],kind="pie")
multi_tickers.max().plot(ax=axes[1,1])
plt.title('Dashboard Showing Top 3 Companies In canada ')

fig.show()
```
